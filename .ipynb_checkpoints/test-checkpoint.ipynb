```python
import Relu
import Dense
import MSE
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1, 10)
x = x.reshape((10, -1))

non_linear_data = x**2 + x*0.1

layer1 = Dense.Dense(1, 100)
layer2 = Dense.Dense(100, 1)
mse = MSE.MSE()
activation = Relu.Relu()

predicts = []
epochs = []

epoch = 100000
for _ in range(epoch):
    # Forward propagation
    z = layer1.forward(x)
    a = activation.forward(z)
    pred = layer2.forward(a)
    
    # Get the current error
    m = mse.calculate(pred, non_linear_data)

    # Backward propagation
    delta = mse.backward(pred, non_linear_data)
    delta = layer2.backward(delta, learning_rate=0.01)
    delta = activation.backward(delta)
    delta = layer1.backward(delta, learning_rate=0.01)

    # Collect pred
    predicts.append(pred)
    epochs.append(_)

plt.scatter(x, predicts[-1])
plt.scatter(x, non_linear_data, color='red')
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
```

```python
x_train, x_test = x_train.reshape((x_train.shape[0], -1)), x_test.reshape((x_test.shape[0], -1))
y_train, y_test = y_train.reshape((-1, 1)), y_test.reshape((-1, 1))
```

```python
import Dense, Relu
hidden_size = 100
layer1 = Dense.Dense(x_train.shape[1], hidden_size)
a_layer1 = Relu.Relu()
output_layer = Dense.Dense(hidden_size, 10)
a_output_layer = Relu.Relu()
```

```python
import MSE
import numpy as np

def accuracy(y_pred, y):
    print(y_pred.shape)
    p = np.argmax(y_pred, axis=1)
    y = y.reshape(-1)
    return np.mean(p == y)

epoch = 11
mse = MSE.MSE()
batch_size = 8

for _ in range(epoch):
    error = 0

    for i in range(0,len(x_train)-batch_size, batch_size):
        
        x = x_train[i:i+batch_size]/255
        y = y_train[i:i+batch_size]
        y = np.eye(10)[y.reshape(-1)]
        # Predict
        z1 = layer1.forward(x)
        a1 = a_layer1.forward(z1)
        z2 = output_layer.forward(a1)
        pred = a_output_layer.forward(z2)
        # Calculate current error
        error += mse.calculate(pred, y)

        # Update weights
        delta = mse.backward(pred, y)
        delta = a_output_layer.backward(delta)
        delta = output_layer.backward(delta,learning_rate=0.01)
        delta = a_layer1.backward(delta)
        layer1.backward(delta,learning_rate=0.01)

    z1 = layer1.forward(x_test/255)
    a1 = a_layer1.forward(z1)
    pred_test_data = a_output_layer.forward(output_layer.forward(a1))
    y_test_one_hot = np.eye(10)[y_test.reshape(-1)]
    test_error = mse.calculate(pred_test_data, y_test_one_hot)
    print(f'epoch {_} train error = {error} Test error = {test_error} Test accuracy = {accuracy(pred_test_data, y_test)}')


    
```

```python
def get_dataset(text, window_size, test_rate = 0.1):
    # Creer le dictionnaire
    chars = sorted(list(set(text)))
    voc_size = len(chars)
    char_to_index = { ch : i for i,ch in enumerate(chars) }
    index_to_char = { i : ch for i,ch in enumerate(chars) }

    # Preparer les sequences brutes
    sequences = []
    next_chars = []
    step = 1

    for i in range(0, len(text) - window_size, step):
        sequences.append(text[i:i+window_size])
        next_chars.append(text[i+window_size])
    
    # Creer X et Y vides
    X = np.zeros((len(sequences), window_size * voc_size), dtype=np.float32)
    Y = np.zeros((len(sequences), voc_size), dtype=np.float32)

    # Remplir les matrices
    for i, sentence in enumerate(sequences):
        for j,c in enumerate(sentence):
            c_index = char_to_index[c]
            col_index = j * voc_size + c_index
            X[i, col_index] = 1.0
            
        target_char = next_chars[i]
        target_index = char_to_index[target_char]
        Y[i, target_index] = 1.0
    test_size = round(0.1 * len(X))
    return X[:-test_size], X[-test_size:], Y[:-test_size], Y[-test_size:] , voc_size, char_to_index, index_to_char

def generate_texte(model, seed_text, c2i, i2c, window_size, voc_size, length=100):
    current_texte = seed_text
    generated = ''

    # Encode current texte
    for _ in range(length):
        x_pred = np.zeros((1, window_size*voc_size))
        for i, c in enumerate(current_texte):
            if (c in c2i):
                c_ind = c2i[c]
                col_ind = c_ind + voc_size * i
                x_pred[:,col_ind] = 1.0
        
        # Predict next char
        next_char_prob = model.predict(x_pred)
        next_char = i2c[np.argmax(next_char_prob)]
    
        # Mise a jour
        generated += next_char
    
        # On decale la fenetre
        current_texte = current_texte[1:] + next_char
        
        print(next_char, end='', flush=True)
    return generated
```

```python
# Preparer le texte
import numpy as np
text = ''
path = "Data/"
file_name = "text.txt"
with open(path+file_name, 'r', encoding='utf-8') as f:
    text = f.read().lower()
print(text[-10:])
window = 20
x_train, x_test, y_train, y_test, voc_size, c2i, i2c = get_dataset(text, window)
x_train.shape
```

```python
import Model
import MSE
import Dense, Relu, Sigmoid
# Preparer le modele
model = Model.Model([
    Dense.Dense(x_train.shape[1], 256),
    Relu.Relu(),
    Dense.Dense(256, voc_size),
    Sigmoid.Sigmoid()],
    MSE.MSE())
model.summary()
```

```python
# train model
model.train(x_train[:60000], y_train[:60000], x_test[:10000], y_test[:10000], 10, 16, learning_rate=0.1)
```
